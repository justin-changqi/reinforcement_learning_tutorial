````python
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import os
````

````python
episodes = 5 
for episode in range(1, episodes+1):
    obs = env.reset()
    done = False
    score = 0
    while not done:
        action = env.action_space.sample()
        obs, reward, done, info = env.step(action)
        score += reward
    print('Episode: {}, Score: {}'.format(episode, score))
env.close()
````

# 3. Train Model
CarRacing Env documentation [Link](https://github.com/openai/gym/blob/master/gym/envs/box2d/car_racing.py)

````python
env = gym.make('CarRacing-v0')
env = DummyVecEnv([lambda: env])
````

````python
log_path = os.path.join('training', 'logs')
model = PPO('CnnPolicy', env, verbose=1, tensorboard_log=log_path)
````

````python
model.learn(total_timesteps=100000)
````

# 4. Save Model

````python
ppo_path = os.path.join('training', 'saved_models', 'PPO_Driving_Model')
````

````python
model.save(ppo_path)
````

````python
del model
````

````python
ppo_path = os.path.join('training', 'saved_models', 'PPO_428k_Driving_model')
````

````python
model = PPO.load(ppo_path, env)
````

# 5. Evaluate and Test

````python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
````
